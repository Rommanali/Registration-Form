```python
import tkinter as tk
from tkinter import ttk
from tkinter import *

window = Tk()
window.title('Registration Form')
window.geometry('600x300')
window.resizable()

def submit():
    print('User filled the form')
    
def value():
    if x.get()==1:
        print('The user is Male')
    else:
        print('The user is Female')

def value2():
    if a.get()==1:
        print('User knows Java')
    else:
        print('User does not know Java')
def value3():
    if b.get()==1:
        print('User knows Python')
    else:
        print('User does not know Python')
    

label = Label(window,text='Registration Form',font=('Serif',25),fg='black',padx=20,pady=15)
label.grid(row=3,column=1)

label1 = Label(window,text='Full name',font=('Serif',11),fg='black')
label1.place(x=20,y=63)

label2 = Label(window,text='Email',font=('Serif',11),fg='black')
label2.place(x=20,y=95)

label3 = Label(window,text='Gender',font=('Serif',11),fg='black')
label3.place(x=20,y=127)


label4 = Label(window,text='Country',font=('Serif',11),fg='black')
label4.place(x=20,y=159)

label5 = Label(window,text='Programming',font=('Serif',11),fg='black')
label5.place(x=20,y=191)

submit = Button(window,text='Submit',font=('Serif',11),fg='white',bg='#1979a9',bd=5,activeforeground='White',
                activebackground='#1979a9',command=submit)
submit.place(x=20,y=245)

entry1= Entry(window,font=('Serif',11),fg='black',width=50)
entry1.place(x=150,y=63)

entry2= Entry(window,font=('Serif',11),fg='black',width=50)
entry2.place(x=150,y=95)

x=IntVar()
Gender = Radiobutton(window,text='Male',font=('Serif',13),fg='black',variable=x,value=1,command=value)
Gender.place(x=155,y=127)
Gender2 = Radiobutton(window,text='Female',font=('Serif',13),fg='black',variable=x,value=2,command=value)
Gender2.place(x=225,y=127)

y=StringVar()
country=ttk.Combobox(window,width=25,textvariable=y,font=('Serif',11))
country['value']=['Select your country','Pakistan','America','China','India','Iran','Afghanistan','Germany','Other']
country.place(x=150,y=159)
country.current(0)

a=IntVar()
java = Checkbutton(window,text='Java',font=('Serif',13),fg='black',variable=a,onvalue=1,offvalue=0,command=value2)
java.place(x=155,y=191)
b=IntVar()
python=Checkbutton(window,text='Python',font=('Serif',13),fg='black',variable=b,onvalue=1,offvalue=0,command=value3)
python.place(x=225,y=191)
window.mainloop()
```
